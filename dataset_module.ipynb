Creating Datasets

```python
from sklearn.datasets import make_blobs
```

```python
# make_blobs = Used to create clusters
# X, y = make_blobs()
# centers = number of clusters
# n_features = dimension of the generated datasets


X, y = make_blobs(n_samples = 1000 , centers = 6 )


X
```

```python
y
```

```python
plt.scatter(X [: , 0] , X[:,1] , c=y)
```

Syntax	         Which Rows?	                                         Which Columns?
X[:, :]	           All rows	                                                All columns
X[0, 0]	       Just the very first row	                             Just the very first column (Single value)
X[:5, 1]	  First 5 rows (0, 1, 2, 3, 4)	                              The second column
