```python
from sklearn.datasets import fetch_openml

from sklearn.preprocessing import OrdinalEncoder
```

```python
data = fetch_openml('car' , as_frame = True).frame
```

```python
data.head(6)
```

Ordinal Encoding  : Used in cases where there is spectrum like which value is higher or lower. Eg: easy , medium , high (Althoough words they show show form  of quantity so they can be easily numbered on order of smallest to largest or lower to higher etc.)

```python
columns_to_encode = ['lug_boot' , 'safety']

encoder = OrdinalEncoder(
    categories = [['small', 'med' , 'big'] , ['low' , 'med' , 'high']]

)

data[columns_to_encode] = encoder.fit_transform(data[columns_to_encode])

```

```python
data.tail(7)
```

```python
data[columns_to_encode] = encoder.inverse_transform(data[columns_to_encode])
```

```python
data.tail()
```

One hot encoding : Used when the label value is not explicitly on any spectrum that indicate higher or lower value. Eg : For race we have white, black , brown etc. since there is no progression (higher or lower) shown by the value only . 

```python
data = fetch_openml('adult' , as_frame = True).frame
```

```python
data.head(4)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
data['occupation'].value_counts()
```

```python
for value in data['occupation'].unique():
    print(value)
```

```python
import pandas as pd

encoder = OneHotEncoder(handle_unknown = 'ignore' , sparse_output = False)

encoded_value  = encoder.fit_transform(data[['occupation', 'race']])

new_column = encoder.get_feature_names_out(['occupation', 'race'])
```

```python
df_encoded = pd.DataFrame(encoded_value , columns = new_column , index = data.index)

data_final = pd.concat([data.drop(columns = ['occupation' , 'race']) , df_encoded] , axis =1 )
```
