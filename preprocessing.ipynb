```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2 , random_state = 17 )
```

Standard Scaler

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train) 
X_test_scaled = scaler.transform(X_test) 
```

X_train_scaled = scaler.fit_transform(X_train)

This code does same thing as :

scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)

We fit and transform in training but only transform in testing becuase fit calculates mean and std deviation so it is like determining the answer first but when testing our model should work from what it has learned rather than what is the correct answer. This helps model to actually learn.
