```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = load_iris()

X, y = data.data , data.target
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, y , test_size = 0.2)
```

```python
counts = np.bincount(y_train)

positions = np.arange(3)


plt.bar(positions , counts)
plt.xticks (positions , data.target_names)
```

Here, after splitting, the test or train data can have irregularities in number of outcomes. For eg : One target value (ground truth) can have much more data than rest (uneven distribution) which can cause problem during prediction.

For eg : Class A has 50 , class B has 40 and class C has 20. So, during prediciton if input belongs to class C it might predict it incorrectly because it is not well generalized for that case. 

To fix this scikit learn provides a method called StartifiedShuffleSplit

```python
data = load_iris()

X, y = data.data , data.target
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split  = StratifiedShuffleSplit(n_splits =  1 , test_size= 0.2)


for train_idx , test_idx in split.split(X , y): 

    X_train , X_test = X[train_idx] , X[test_idx]
    y_train , y_test = y[train_idx] , y[test_idx]
```

```python
counts = np.bincount(y_train)

positions = np.arange(3)


plt.bar(positions , counts)
plt.xticks (positions , data.target_names)
```

```python
counts = np.bincount(y_test)

positions = np.arange(3)


plt.bar(positions , counts)
plt.xticks (positions , data.target_names)
```
